```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import KBinsDiscretizer
```

```python
df = pd.read_csv(r'C:\Users\maype\OneDrive - sga.pucminas.br\Área de Trabalho\Faculdade\projects\A Machine Learning Approach to Characterizing Depression in Brazilian Adult\Data\Bases Modificadas\correlacoes_entopia_removido.csv')

df.head()
```

```python
df.columns 
```

```python
# Aplica o One-Hot Encoding
df = pd.get_dummies(df, columns=['Sexo'], prefix='Sexo')
df.rename(columns={'Sexo_1.0': 'sexo_masculino', 'Sexo_2.0': 'sexo_feminino'}, inplace=True)
df[['sexo_masculino','sexo_feminino']].value_counts()
```

### Discretizando/Codificação Dados

```python
# Definindo as faixas para discretização
faixas_renda = [0, 1045, 2090, 4180, 8360, 16720, np.inf]  # Faixas de renda do IBGE (2022)
faixas_imc = [0, 18.5, 24.9, 29.9, 34.9, 39.9, np.inf]  # Faixas de IMC da OMS
```

```python
# Criando o objeto KBinsDiscretizer
discretizador_renda = KBinsDiscretizer(n_bins=len(faixas_renda)-1, encode='ordinal', strategy='uniform')
discretizador_imc = KBinsDiscretizer(n_bins=len(faixas_imc)-1, encode='ordinal', strategy='uniform')

# Discretizando as colunas
df['renda_discretizada'] = discretizador_renda.fit_transform(df[['renda_total']]).astype(int)
df['imc_discretizado'] = discretizador_imc.fit_transform(df[['IMC']]).astype(int)
```

```python
df[['renda_discretizada', 'imc_discretizado']].head()
```

```python
df.drop(columns=['IMC', 'renda_total', 'peso_consolidado(KG)', 'altura_consolidada(CM)'], inplace=True)
df.head()
```
