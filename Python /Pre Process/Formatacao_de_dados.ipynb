### Criando Resumo 
> Agora vamos criar uma tabela que mostre cada variável, seu tipo e a quantidade de valores alsentes. 

```python
# Função para resumir as informações
def resumir_variaveis(df):
    resumo = pd.DataFrame({
        'Variável': df.columns,
        'Tipo': [df[col].dtype for col in df.columns],
        'Valores Nulos': [df[col].isnull().sum() for col in df.columns]
    })
    return resumo

# Gerando a tabela de resumo
resumo_variaveis = resumir_variaveis(df)


# Exibindo a tabela de resumo (opcional)
print(resumo_variaveis)
```
